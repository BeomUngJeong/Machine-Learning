```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=2*np.random.rand(100,1)
temp_x=x
y=4+3*x+np.random.randn(100,1)
temp_y=y
```

```python
lr=0.1
iter=100
W=np.random.randn(2,1)
col0=np.ones((100,1))
col1=x
x=np.concatenate((col0,col1),axis=1)
```

```python
y_hat=np.dot(x,W)
```

```python
for i in range(100):
    gradient_w0=W[0][0]+W[1][0]*x[i][0]-y[i][0]
    gradient_w1=x[i][0]*(W[0][0]+W[1][0]*x[i][0]-y[i][0])
    W[0][0]-=lr*gradient_w0
    W[1][0]-=lr*gradient_w1
    print(W)
```
