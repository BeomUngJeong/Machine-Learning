```python
import numpy as np
# np.arange, np.asarray, np.random.rand
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# objective function, derivatice function
def objective_func(x):
    return x**2
def derivative_func(x):
    return x*2

# input and target
x_min,x_max=-1.0,1.0
inputs=np.arange(x_min,x_max+0.1,0.1)
targets=objective_func(inputs)

#visualization
plt.plot(inputs,targets)
plt.show()
```

```python
# gradient descent algorithm
def gradient_descent(objective_func, derivative_func, bounds = np.asarray([[-1.0, 1.0]]), n_iter=30, step_size=0.1):
    solutions, scores = list(), list()
    solution = bounds[:, 0] + np.random.rand(len(bounds)) * (bounds[:, 1] - bounds[:, 0])
    for i in range(n_iter):
        gradient = derivative_func(solution)
        solution = solution - step_size * gradient
        solution_eval = objective_func(solution)
        solutions.append(solution)
        scores.append(solution_eval)
        print('>%d f(%s) = %.5f' % (i, solution, solution_eval))
    return [solutions, scores]

solutions, scores = gradient_descent(objective_func, derivative_func)
inputs = np.arange(bounds[0,0], bounds[0,1]+0.1, 0.1)
results = objective_func(inputs)
```

```python
plt.plot(inputs, results)
plt.plot(solutions, scores, '.-', color='red')
plt.show()
```
