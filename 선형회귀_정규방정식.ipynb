```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(0,1,101)
col0=np.ones((101,1))
col1=x.reshape(-1,1)
col2=(x*np.random.random(len(x))).reshape(-1,1)

y=x*np.random.random(len(x)) + (x+1)

x=x.reshape(-1,1)
y=y.reshape(-1,1)
```

```python
X=np.concatenate([col0,col1,col2],axis=1)
```

```python
#정규방정식
W=np.dot(np.dot(np.linalg.inv(np.dot(X.T,X)),X.T),y)
W=W.ravel()
```

```python
#Curve Fitting
plt.scatter(x,y)
plt.plot(x,W[0]+W[1]*x)
```
